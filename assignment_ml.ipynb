Ques 2->What are the key assumptions of Simple Linear Regression?

Ans->The key assumptions of Simple Linear Regression are:

1-Linearity
The relationship between the independent variable (X) and dependent variable (Y) is linear.

2-Independence of errors
The residuals (errors) are independent of each other.

3-Homoscedasticity
The variance of residuals is constant for all values of X.

4-Normality of errors
The residuals are normally distributed (mainly important for hypothesis testing).

5-No significant outliers
There are no extreme outliers that strongly influence the regression line.

Ques 3->What is heteroscedasticity, and why is it important to address in regression
models?

Ans->Heteroscedasticity occurs in a regression model when the variance of the error terms (residuals) is not constant across all values of the independent variable(s).

      *Why it is important to address*

1-Invalid statistical test:
Standard errors become biased, making t-tests and F-tests unreliable.

2-Incorrect confidence intervals:
Confidence intervals may be too wide or too narrow.

3-Misleading significance of predictors:
Variables may appear significant when they are not (or vice versa).

4-Reduced model efficiency:
Estimates are no longer the most efficient (not minimum variance).

Ques 6->Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
â— X = [1, 2, 3, 4, 5]
â— Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.

```python
#Ans 6->import numpy as np

import matplotlib.pyplot as plt
import numpy as np

X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Fit Simple Linear Regression
m, c = np.polyfit(X, Y, 1)

# Predicted values
Y_pred = m * X + c

plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()
```

Ques 7-> Fit a Multiple Linear Regression model on this sample data:
â— Area = [1200, 1500, 1800, 2000]
â— Rooms = [2, 3, 3, 4]
â— Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.


```python
#Ans 7->

import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Given data
data = {
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[["Area", "Rooms"]]
y = df["Price"]

# Add constant term
X_const = sm.add_constant(X)

# Fit Multiple Linear Regression model
model = sm.OLS(y, X_const).fit()

# Calculate VIF
vif = pd.DataFrame()
vif["Feature"] = X_const.columns
vif["VIF"] = [
    variance_inflation_factor(X_const.values, i)
    for i in range(X_const.shape[1])
]

print(model.summary())
print("\nVIF Values:")
print(vif)
```

 VIF values between 5 and 10 indicate moderate multicollinearity.

Area and Rooms are somewhat correlated, but not severely.

The model is usable, but multicollinearity should be monitored if more data or variables are added.

Ques 8->Implement polynomial regression on the following data:
â— X = [1, 2, 3, 4, 5]
â— Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve


```python
#Ans 8->
import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Fit a 2nd-degree polynomial
coeffs = np.polyfit(X, Y, 2)
poly_model = np.poly1d(coeffs)

# Generate smooth curve
X_smooth = np.linspace(X.min(), X.max(), 100)
Y_smooth = poly_model(X_smooth)

# Plot data points and polynomial curve
plt.scatter(X, Y)
plt.plot(X_smooth, Y_smooth)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.show()
```

Ques 9-> Create a residuals plot for a regression model trained on this data:
â— X = [10, 20, 30, 40, 50]
â— Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
X = np.array([10, 20, 30, 40, 50])
Y = np.array([15, 35, 40, 50, 65])

# Fit simple linear regression
m, c = np.polyfit(X, Y, 1)
Y_pred = m * X + c

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals)
plt.axhline(0)
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot for Heteroscedasticity Check")
plt.show()
```
